```python
import pandas as pd
from sqlalchemy import create_engine
from datetime import datetime as dt
```

```python
# upload  anime csv file
csv_file = "./Resources/Anime.csv"
anime_df = pd.read_csv(csv_file)
anime_df.head()
```

```python
# pull desired columns 
# drop nulls in any of those columns and reset the index
cleaned_anime_df = anime_df[['Name','Tags','Release_year','Type']].dropna().reset_index()
cleaned_anime_df.head()
```

```python
# convert the year from int to date time 
# make date time objects first
cleaned_anime_df['Release_year'] = pd.to_datetime(cleaned_anime_df['Release_year'], format = '%Y')
# then extract out the year
cleaned_anime_df['Release_year'] = cleaned_anime_df['Release_year'].dt.year
```

```python
# renname the columns
cleaned_anime_df = cleaned_anime_df.rename(columns={'Name':'Title'})
cleaned_anime_df = cleaned_anime_df.drop_duplicates()
```

```python
# convert to desired data types
cleaned_anime_df['Title'] = cleaned_anime_df['Title'].astype(str)
cleaned_anime_df['Tags'] = cleaned_anime_df['Tags'].astype(str)
cleaned_anime_df['Release_year'] = cleaned_anime_df['Release_year'].astype(int)
cleaned_anime_df.head()
```

```python
# upload  netflix csv file
csv_file = "./Resources/NetflixOriginals.csv"
netflixog_df = pd.read_csv(csv_file, encoding = "ISO-8859-1")
netflixog_df.head()
```

```python
new_netflix_df = netflixog_df[['Title','Genre','Premiere']].dropna().reset_index()
new_netflix_df = new_netflix_df.rename(columns={'Genre':'Tags','Premiere': 'Release_year'})
new_netflix_df = new_netflix_df.drop_duplicates()
new_netflix_df.head()
```

```python
# there is an error where the format is month day. year instead of month day,year
# this will catch and convert them so that they can be converted with the date time formatter later
for row in new_netflix_df.iterrows():
    initial_value = row[1]['Release_year']
    if row[1]['Release_year'].find('.')!=-1:
        new_value = row[1]['Release_year'].replace('.',',')
        new_netflix_df['Release_year'] = new_netflix_df['Release_year'].replace({initial_value:new_value})
```

```python
# convert to datetime object
new_netflix_df['Release_year'] = pd.to_datetime(new_netflix_df['Release_year'], format = '%B %d, %Y')
```

```python
# convert to desired data types
new_netflix_df['Title'] = new_netflix_df['Title'].astype(str)
new_netflix_df['Tags'] = new_netflix_df['Tags'].astype(str)
new_netflix_df['Release_year'] = new_netflix_df['Release_year'].dt.year.astype(int)
new_netflix_df.head()
```

```python
# add the type column 
new_netflix_df['Type'] = 'Netflix'
new_netflix_df.head()
```

```python
# upload US/Youtube csv file
csv_file = "./Resources/USvideos.csv"
usvideos_df = pd.read_csv(csv_file)
usvideos_df.head()
```

```python
# new df
yt_usvideos = usvideos_df[['title', 'tags','publish_time']]
yt_usvideos = yt_usvideos.rename(columns={'title':'Title','tags':'Tags','publish_time': 'Release_year'})
yt_usvideos.head()
```

```python
# use date time to get just the year
yt_usvideos['Release_year'] = pd.to_datetime(yt_usvideos['Release_year'], format = '%Y-%m-%d %H:%M:%S')
yt_usvideos['Release_year'] = yt_usvideos['Release_year'].dt.year
yt_usvideos.head()
```

```python
yt_usvideos['Type'] = 'YT Video'
yt_usvideos = yt_usvideos.drop_duplicates()
yt_usvideos.head()
```

```python
# convert to desired data types
yt_usvideos['Title'] = yt_usvideos['Title'].astype(str)
yt_usvideos['Tags'] = yt_usvideos['Tags'].astype(str)
yt_usvideos['Release_year'] = yt_usvideos['Release_year'].astype(int)
```

```python
# join them into a full table 
all_media_df = cleaned_anime_df.append(new_netflix_df)
all_media_df = all_media_df.append(yt_usvideos) 
all_media_df.head()
```

```python
all_media_df = all_media_df.drop('index', 1)
# drop duplicates in all table 
all_media_df = all_media_df.drop_duplicates()
```

```python
# connect to local database 
rds_connection_string = "postgres:postgres@localhost:5432/media"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# use pandas to load csv converted DF into DB
# cleaned versions
cleaned_anime_df.to_sql(name='Anime', con=engine, if_exists='append', index=False)
new_netflix_df.to_sql(name='Netflix', con=engine, if_exists='append', index=False)
yt_usvideos.to_sql(name='Youtube/US', con=engine, if_exists='append', index=False)
all_media_df.to_sql(name='All_Media', con=engine, if_exists='append', index=False)
```
